<h2><center>Part Two: Steady Rotation

<strong><font color=green> Problem 9.12: As you perform the steady rotation, record the angle and galvanometer deflection (sign and magnitude) on the table below.

| | 3 o'clock | 6 o'clock | 9 o'clock | 12 o'clock|
|----|----|----|----|----|
|Current [$\mu A$]| 0 | 300 | 0 | -300 |

```python
current_measurements = [0,300,0,-300]
angles = [0,3/2*np.pi,np.pi,np.pi/2]

def oscillatory(angle,amplitude,phase):
    y = amplitude*np.cos(angle+phase)
    return y

params, ecov = optimize.curve_fit(oscillatory, angles, current_measurements)

angles_lin = np.linspace(0,2*np.pi,360)
current_lin = oscillatory(angles_lin,params[0],params[1])


plt.figure()
plt.plot(angles_lin,current_lin,'k-.')
plt.plot(angles,current_measurements,'go')
plt.ylabel('Current [$\mu$A]')
plt.xlabel('Angles [Radians]')
plt.title('Angle Position vs. Galvanometer Current')
plt.show()
```
