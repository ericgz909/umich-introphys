As we mentined before, we will use `matplotlib.pyplot` to make plots. We can make a plot of the coulomb force as a function of distance between two charges $Q$ and $-Q$<font color=white><#!S!#>

```python
# define charges
Q1 = 1.
Q2 = -1.

# define electric constant
epsilon = 8.85e-12
k_e = 1./(4*np.pi*epsilon)

# define a function to calculate the force
def coulomb_force(r,q1,q2):
    F = k_e * q1 * q2 /(r**2)
    return F

# pick radii to plot
radii = np.array([1,2,3,4,5])
# or we can do a continuous array with linspace
# np.linspace(start,stop,step) gives an array from 'start' to 'stop' with step size 'step'
radii_cont = np.linspace(1,5,1000)

force = coulomb_force(radii,Q1,Q2)
force_cont = coulomb_force(radii_cont,Q1,Q2)

# define a plt.figure, a canvas to draw plots on
plt.figure(figsize=(16,10))

# divide into two subplots using plt.subplot(nml) 
# where n is the number of rows, m the columns, and l the subplot index
# for example, to plot two plots in a row, we let n=1, m=2 
# and we specify the plot we are currently drawing on with l.

# starting with the first plot, a discrete plot of coulomb force
plt.subplot(121)
# plot using plt.plot(x,y) - there are many other plotting functions e.g. scatter, histogram, etc.
plt.plot(radii,force)

# then next to it, we switch to the second plot l=2 and plot the continuous values
plt.subplot(122)
plt.plot(radii_cont,force_cont)

# finally we call plt.show() to show the plot below
plt.show()
```

<strong>Clearly the continuous sampling is a much more accurate representation of the coulomb force. There is one major problem with these plots, they are completely unlabeled. We can use the following functions to label our plots
